```python
dataset1=pd.read_csv("prep.csv",index_col=None)
df2=dataset1
df2 = pd.get_dummies(df2, drop_first=True)

X=df2.drop('classification_yes', axis=1)
y=df2['classification_yes']
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(solver='lbfgs')
sfs1 = SequentialFeatureSelector(log_model,k_features=10, forward=True, floating=False, verbose=0, scoring=None, cv=5, n_jobs=1, pre_dispatch='2n_jobs', clone_estimator=True, fixed_features=None, feature_groups=None)
```

```python
sfs1 = sfs1.fit(X, y)
sfs1
```

```python
sfs2=sfs1.transform(X)
sfs2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sfs2, Dep_Y, test_size = 0.25, random_state = 0)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
        
        # Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
        
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
        
Accuracy=accuracy_score(y_test, y_pred )
        
report=classification_report(y_test, y_pred)
```
